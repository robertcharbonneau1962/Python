```python
#911 Calls

df = pd.read_csv('911.csv')
df.head()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
len(df['title'].unique())
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
df['Reason'].value_counts().head()
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['timeStamp']
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
time.month
```

```python
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of the week'] = df['Day of week'].map(dmap)
df['Day of the week']
```

```python
sns.countplot(x='Day of week',data=df,hue='Reason',palette='viridis')
```

```python
sns.countplot(x='Month',data=df,palette='viridis')
```

```python
t = df['timeStamp'].iloc[0]
t.date()
```

```python
df['Date'] = df['timeStamp'].apply(lambda t : t.date())
df['Date'].head()
```

```python
# BANKS

# IMPORTS TO READ STOCKS
import datetime
from pandas_datareader import data,wb
```

```python
df = pd.read_pickle('all_banks')
```

```python
df.info()
```

```python
df['BAC'].head()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']  #show name on top of table
bank_stocks = pd.concat([df['BAC'],df['C'],df['GS'],df['JPM'],df['MS'],df['WFC']],axis=1,keys=tickers)
bank_stocks.head()
```

```python
#FOR EACH
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
returns = pd.DataFrame()
for tick in tickers:
    returns[tick+' Returns'] = bank_stocks[tick]['Close'].pct_change()

returns.head()
```

```python
sns.pairplot(returns[1:])
```

```python
returns.idxmin()
```

```python
returns.std()
```

```python
returns.loc['2015-01-01':'2015-12-31'].std() #STD based on specific dates
```

```python
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Returns'],color='green',bins=50)
```

```python
sns.displot(returns.loc['2008-01-01':'2008-12-31']['C Returns'],color='red',bins=50)
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(label=tick,figsize=(12,4))
plt.legend()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').plot()
```

```python
#Stock MA
plt.figure(figsize=(12,4))
bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 MA')
bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
