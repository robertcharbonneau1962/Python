```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x= np.linspace(0,5,11)
y = x ** 2
plt.plot(x,y)
plt.xlabel('y_label')
plt.ylabel('y_label')
plt.title('Title')
#test
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1 plot')

axes[1].set_title('2 plot')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X_squared')
ax.plot(x,x**3,label='x_cubed')

ax.legend(loc=1)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])

ax1.plot(x,y,color='red')
ax2.plot(x,y,color='green')
```
